```python
import numpy as np

data = np.array([100, 200, 300, 400])

def data_manipulation(data, lowest_light_level):
    L1, L2, L3, L4 = data[0], data[1], data[2], data[3]
    avg_left, avg_right, avg_up, avg_down = np.mean([L1, L3]), np.mean([L2, L4]), np.mean([L1, L2]), np.mean([L3, L4])
    diff_vert = avg_up - avg_down
    diff_horiz = avg_right-avg_left

    if diff_vert > 0:
        print("Rotate up")
    elif diff_vert == 0:
        print("Do not rotate vertically")
    else:
        print("Rotate down")

    if diff_horiz > 0:
        print("Rotate right")
    elif diff_horiz == 0:
        print("Do not rotate horizontally")
    else:
        print("Rotate left")
        
    ## Sum light levels
    if sum([L1, L2, L3, L4]) <= lowest_light_level:
        print(f"Turn off module - sum is at {sum([L1, L2, L3, L4])}")
    else:
        print(f"Keep module on - sum is at {sum([L1, L2, L3, L4])}")

    ## Rough idea of the accuracy - the higher the better
    mean_brightness = np.mean([L1, L2, L3, L4])

    return diff_vert, diff_horiz, mean_brightness

## DIff_vert: up is positive, down is negative
## Diff_horiz: right is positive, left is negative
data_manipulation(data, 100)
```
